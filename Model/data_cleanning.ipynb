```python
import sys
sys.path.append('..')

import pandas as pd
from Model.preprocessing_util import *
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

def preprocess_data():
    # csv파일 열기
    df = pd.read_csv('../Data/hanacard_sample.csv')
    
    #필요 없는 특성 제거
    allColumns = df.columns
    params = ['기준년월','광역시도명','시군구명','읍면동명','당사업종대분류코드명','당사업종중분류코드명','당사업종코드','업종명','구분_개인법인','구분_연령','구분_성별','구분_직장','추정소득구간','신용등급구간','카드이용건수','카드이용금액']

    df = df.drop(allColumns.drop(params), axis=1)
    df = df.dropna()
    
#     df['기준년월']
#     df['광역시도명']
#     df['시군구명']
#     df['읍면동명'] 
#     df['당사업종대분류코드명']
#     df['당사업종중분류코드명']
#     df['당사업종코드']
#     df['업종명']
#     df['구분_개인법인']
    df['구분_연령'] = df['구분_연령'].apply(age_to_int)
#     df['구분_성별']
#     df['구분_직장']
    df['추정소득구간'] = df['추정소득구간'].apply(str_to_median)
    df['신용등급구간'] = df['신용등급구간'].apply(str_to_median)
#     df['카드이용건수']
#     df['카드이용금액']

    # one-hot 인코딩
    ct = ColumnTransformer([
    ('onehot', OneHotEncoder(sparse=False), ['당사업종대분류코드명','당사업종중분류코드명','업종명','구분_개인법인','구분_성별','구분_직장'])], remainder='passthrough'
    )
    ct.fit(df)
    X = ct.transform(df)
    
    # 컬럼 이름 리스트 생성
    num_cols = df.columns.tolist()
    ohe = ct.named_transformers_['onehot']
    ohe_cols = ohe.get_feature_names_out(['당사업종대분류코드명','당사업종중분류코드명','업종명','구분_개인법인','구분_성별','구분_직장']).tolist()
    new_cols = ohe_cols + num_cols
    new_cols.remove('당사업종대분류코드명')
    new_cols.remove('당사업종중분류코드명')
    new_cols.remove('업종명')
    new_cols.remove('구분_개인법인')
    new_cols.remove('구분_성별')
    new_cols.remove('구분_직장')
    
    # DataFrame으로 변환
    df = pd.DataFrame(X, columns=new_cols)
    
    # 최대 열 및 행 제한 해제
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)
    print(df)
    
preprocess_data()
```
